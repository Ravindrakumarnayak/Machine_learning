```python
# to develop into unique structure, we use power pivot


user_movie_df=movie_df.pivot( index='userId',# rows
                             columns='movie',#columns
                              values='rating')
```

```python
user_movie_df
```

```python
user_movie_df.fillna(0, inplace=True) # fillna method will convert all NaN to O and inplace true means in the same dataframe convert it
```

```python
user_movie_df
```

```python
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine,correlation
```

```python
user_sim= 1- pairwise_distances(user_movie_df.values, metric='cosine')
```

```python
user_sim # similarity matrix
```

```python
np.fill_diagonal(user_sim,0)
```

```python
user_sim
```

```python
user_sim_df=pd.DataFrame(user_sim)
```

```python
user_sim_df
```

```python
user_sim_df.index=movie_df.userId.unique()
user_sim_df.columns=movie_df.userId.unique()
```
