```python
from sklearn.manifold import TSNE
from bioinfokit.visuz import cluster

import warnings
warnings.filterwarnings("ignore")
```

```python
df= pd.read_csv("/content/TSNE_data.csv")
df
```

```python
df.head() #( M = malignent= having cancer), (B= benign= no cancer)
```

```python
# putting all values in array dataset and then  x is independent variable and y is dependent variable
array =df.values

x=array[:,1:]
y= array[:,0]
```

```python
df
```

```python
data_tsne= TSNE(n_components=2).fit_transform(x)
```

```python
data_tsne
```

```python
new_df=pd.DataFrame(data_tsne).join(df['diagnosis'])
new_df
```

```python
# here get plot for those who are concer patient and those who are not...

color_class= df['diagnosis'].to_numpy()

cluster.tsneplot(score= data_tsne, colorlist= color_class, legendpos= 'upper right')

plt.show()
```
