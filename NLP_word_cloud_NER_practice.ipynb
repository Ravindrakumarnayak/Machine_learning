```python
nlp= spacy.load("en_core_web_sm")
```

```python
nlp.pipe_names
```

```python
doc= nlp(" Mr. Elon is acquaring  Twitter Inc at $ 450000 in 2023")
```

```python

doc.ents
```

```python
for i in doc.ents:
  print(i,"|", i.label_)
```

```python
doc1= nlp("  Elon is acquaring  Twitter Inc at $ 450000 in 2023")
```

```python
for i in doc1.ents:
  print(i,"|", i.label_)
```

```python
# spacy is 100%accurate
```

```python
#displacy for graphical representation
```

```python
displacy.render(doc,style='ent', jupyter=True)
```

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
```

```python
stop= STOPWORDS
```

```python
wc=WordCloud(background_color='White', stopwords=stop).generate(txt) # wordcloud will generate bigger text for important words,frequently used word
plt.imshow(wc)
```
